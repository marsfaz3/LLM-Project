```python
from datasets import load_from_disk
ds = load_from_disk("/content/drive/MyDrive/Colab-Notebooks/LLM-Project/datasets/20_newsgroups_ds")
ds
```

```python
import pandas as pd
ds_train = pd.DataFrame(ds['train'])
ds_test = pd.DataFrame(ds['test'])
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="alocken/newsgroups_model")
```

```python
preds = pipe(ds_train['text'].tolist(), truncation=True)
preds
```

```python
from sklearn.metrics import classification_report

y_true = ds_train['label']
y_pred = [int(pred['label'].split('_')[-1]) for pred in preds]

print(classification_report(y_true, y_pred))
```
